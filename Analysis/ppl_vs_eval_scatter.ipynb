```python
for lang in chosen_languages:
    for model in models:
        for template in templates:
            df = pd.read_csv(f'{dir_path}/{lang}_{model}_{template}.csv')
            ls_acc_norm = df['acc_norm'].tolist()
            ls_byte_perplexity = df['byte_perplexity'].tolist()
            ls_acc = df['acc'].tolist()
            ls_word_perplexity = df['word_perplexity'].tolist()

            plt.scatter(ls_byte_perplexity, ls_acc_norm, color='blue', s=100, label='Sample Data')


            # Add title and labels
            plt.title(f'{lang} - {model} - {template}')
            plt.xlabel('X Axis')
            plt.ylabel('Y Axis')

            # Add a legend
            plt.legend()

            # Show the plot
            plt.show()
```
