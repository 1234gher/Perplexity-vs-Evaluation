```python
import pandas as pd
```

```python
df = pd.read_csv('Stat Test Results - Language Results.csv')
```

```python
column_names = df.columns
```

```python
rows_1 = df.iloc[0].to_dict()
```

```python
import numpy as np
```

```python
len(rows_1)
```

```python
prev = None
col_name = []
rows_1_keys = list(rows_1.values())
col_name.append(rows_1_keys[0])
col_name.append(rows_1_keys[1])
for k, v in rows_1.items():
    k = k.strip()
    if str(v) == 'nan':
        col_name.append('nan')
        continue
    if 'unnamed' in k.lower():
        if prev is not None:
            col_name.append(prev + ' - ' + v)
            print(f'{prev} = {v}')
        # else:
            # col_name.append(k)
            # continue
    else:
        prev = k
        col_name.append(k + ' - ' + v)
        print(f'{k} = {v}')
```

```python
# set col_name as column names
df.columns = col_name
```

```python
# drop first row
df = df.drop(0)
```

```python
df
```

```python
mw_test_word_perplexity = df['WORD PERPLEXITY - MW Is significant'].to_list()
mw_test_byte_perplexity = df['Byte Perplexity - MW Is significant'].to_list()
mw_test_log_word_perplexity = df['Log Word Perplexity - MW Is significant'].to_list()

# 'TRUE' is True and 'FALSE' is Fals
mw_test_word_perplexity = [True if x == 'TRUE' else False for x in mw_test_word_perplexity]
mw_test_byte_perplexity = [True if x == 'TRUE' else False for x in mw_test_byte_perplexity]
mw_test_log_word_perplexity = [True if x == 'TRUE' else False for x in mw_test_log_word_perplexity]

import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

# Sample boolean lists
list1 = mw_test_word_perplexity
list2 = mw_test_byte_perplexity
list3 = mw_test_log_word_perplexity

# Convert boolean lists to sets of indices with True values
set1 = {i for i, x in enumerate(list1) if x}
set2 = {i for i, x in enumerate(list2) if x}
set3 = {i for i, x in enumerate(list3) if x}

print(set1)
print(set2)
print(set3)

from venny4py.venny4py import *

#dict of sets
sets = {
    'Word Perplexity': set1,
    'Byte Perplexity': set2,
    'Log Word Perplexity': set3
}
    
venny4py(sets=sets)
```
